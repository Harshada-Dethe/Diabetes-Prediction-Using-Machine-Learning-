# DESCRIPTION : 

The objective of this dataset is to predict whether patient has diabetes or not. The datasets consists of several medical predictor (Independent) variables and one target variable,(Outcome).Predicted variables includes Pregancies, Glucose, Blood Pressure, Skin Thickness, Insulin, BMI, DiabetesPedigreeFunction, age and Outcome.

Dataset url: https://www.kaggle.com/uciml/pima-indians-diabetes-database

# Name of the columns with their meanings-

1) Pregnancies ---> Number of times pregnant                                                                                    
2) Glucose ---> Plasma glucose concentration on 2 hours in an oral glucose tolerance test                                       
3) BloodPressure ---> Diastolic blood pressure (mm Hg)                                                                           
4) SkinThickness ---> Triceps skin fold thickness (mm)                                                                           
5) Insulin ---> 2-Hour serum Insulin (mu U/ml)                                                                                   
6) BMI --->Body mass index (weight in kg/(height in m)^2)                                                                       
7) DiabetesPedigreeFunction ---> Diabetes pedigree function                                                                     
8) Age ---> Age (years)                                                                                                         
9) Outcome ---> Class variable (0 or 1)                                                                                                            268 of 768 are 1     ,                                                                                                                                     500 of 768 are 0

# ROC curve And ROC AUC

ROC curve is one of the important evaluting metrics that should be used to check the perfomance of classification model. It is also called relative operating characteristic curvr, because it is a comparison of two main characteristics (TPR and FPR), It is plotted between sensitivity (recall True Positive Rate) and False Positive Rate (FTR = 1-specificity).

ROC (Receiver Operating Characteristic) Curve tells us about How Good the Model can distinguish between two things (eg.If a patient has a dieses or Not).

Area Under Curve (AUC) helps us to choose the Best Model among the Models for which we have plotted the ROC curves

```python
# Area Under Curve
auc = roc_auc_score(y_test, lr_pred)
print('RDC AUC SCORE of Logistic Regression is',auc)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, lr_pred)
plt.plot(fpr, tpr, color='orange', label='ROC')
plt.plot([0,1],[0,1], color='darkblue', linestyle='--', label='ROC curve (area = %0.2f)' % auc)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve of Logistic Regression')
plt.legend()
plt.grid()
plt.show()
```
